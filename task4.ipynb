### level 3

```python
total_revenue_region = df.groupby('Region')['Total'].sum()
print(total_revenue_region)
```

```python
avg_sale_salesperson = df.groupby('Salesperson')['Total'].mean()
print(avg_sale_salesperson)
```

```python
orders_per_customer_type = df['CustomerType'].value_counts()
print(orders_per_customer_type)
```

```python
revenue_per_product = df.groupby('Product')['Total'].sum()
highest_revenue_product = revenue_per_product.max()
print("\nProduct with highest total revenue:", highest_revenue_product)
```

```python
units_per_salesperson = df.groupby('Salesperson')['Units'].sum()
top_salesperson_units = units_per_salesperson.idxmax()
print("\nSalesperson who sold most units:", top_salesperson_units)
print(units_per_salesperson)
```

### Level 4

```python
df['Month'] = df['Date'].dt.month
```

```python
df
```

```python
sales_per_month = df.groupby('Month')['Total'].sum().sort_values(ascending=False)
print("Total sales per month:")
print(sales_per_month)
```

```python
highest_month = sales_per_month.idxmax()
print("\nMonth with highest revenue:", highest_month)
```

```python
avg_order_per_month = df.groupby('Month')['Total'].mean()
print("\nAverage order value per month:")
print(avg_order_per_month)
```

```python
summary_region_month = df.groupby(['Region', 'Month'])['Total'].sum()
print("\nSummary of total sales by region and month:")
print(summary_region_month)
```

### level 5 

```python
top3_products = df.groupby('Product')['Total'].sum().sort_values(ascending=False).head(3)
print("Top 3 products by revenue:")
print(top3_products)
```

```python
total_revenue = df['Total'].sum()
salesperson_contribution = (df.groupby('Salesperson')['Total'].sum() / total_revenue) * 100
print("\nSalesperson contribution (% of total revenue):")
print(salesperson_contribution)
```

```python
df['DiscountedTotal'] = df['Total'] * 0.95
print("\nNew column 'DiscountedTotal' added.")
```

```python
most_sold_per_region = df.groupby(['Region', 'Product'])['Units'].sum().reset_index()
most_sold_per_region = most_sold_per_region.loc[most_sold_per_region.groupby('Region')['Units'].idxmax()]
print("\nMost sold product per region:")
print(most_sold_per_region)
```
